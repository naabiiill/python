```python
#Given List of fruits
fruits = ["apple", "banana", "cherry", "mango"]

# Printing index and fruit name
for index, fruit in enumerate(fruits):
    print(index, fruit)
```

```python
# Lists of students and scores
students = ["Rahim", "Karim", "Ayesha"]
scores = [85, 90, 78]

# Print each student with their score
for student, score in zip(students, scores):
    print(f"{student} scored {score}")
```

```python
# Tuple of subjects
subjects = ("Math", "Physics", "Chemistry", "English", "Biology")

# Print each subject
for subject in subjects:
    print(subject)
```

```python
# Given sets
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7}

# Print common elements
print("Common elements:", set1 & set2)

# Print unique elements of set1
print("Unique elements of set1:", set1 - set2)
```

```python
# Dictionary of marks
marks = {
    "Math": 80,
    "Physics": 75,
    "Chemistry": 85
}

# Print subject number, name, and marks
for index, (subject, mark) in enumerate(marks.items(), start=1):
    print(f"{index} {subject}: {mark}")
```
