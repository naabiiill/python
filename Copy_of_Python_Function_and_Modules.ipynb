## Introduction to Functions

**A function is a reusable block of code that performs a specific task.**

🧠 Think of it like a machine:

- You give it some input

- It does something

- It gives back output


💡 so, Function is kind of like a machine (let’s say a coffee machine) which can take input (e.g., coffee beans, milk, sugar, etc.), then process it (grind the beans, heat the water, mix everything together), and finally give some output (a cup of coffee).

````python
# Scenario: An e-commerce system needs to calculate final price of products

# -------------------------------
# ❌ BAD PRACTICE: Repeating code
# -------------------------------

# Case 1
price = 1000       # oven
tax_rate = 0.25    # 25%
total_price = price + (price * tax_rate)   # logic / calcultion -> final price
print("Final Price:", total_price)


# Case 2 (same logic repeated)
price = 2000     # tv
tax_rate = 0.25
total_price = price + (price * tax_rate) # logic / calcultion -> final price
print("Final Price:", total_price)


# Case 3 (same logic repeated again)
price = 500     # t-shirt
tax_rate = 0.25
total_price = price + (price * tax_rate)  # logic / calcultion -> final price
print("Final Price:", total_price)



# ❗ Problem:
# - Code duplication
# - Hard to maintain
# - If tax rate changes, you must update everywhere
````

````python
# ============================================================
# ✅ GOOD PRACTICE: Turn repeated logic into a FUNCTION
# ============================================================

def calculte_final_price(price):
    tax_rate = 0.25
    total_price = price + (price * tax_rate)  # logic / calcultion -> final price
    print("Final Price:", total_price)
````

````python
# -------------------------------
# Using the function (Reusable!)
# -------------------------------

calculte_final_price(3000)   # shirt
calculte_final_price(2490)   # panjabi
calculte_final_price(3400)   # pant
calculte_final_price(220)    # t-shirt
````

````python
# ============================================================
# ✅ Why this is better (Function Usability)
# ============================================================

# 1️⃣ Reusability:
#    Same logic used multiple times without rewriting code

# 2️⃣ Maintainability:
#    Change tax rate in ONE place only

# 3️⃣ Readability:
#    calculate_final_price() clearly explains WHAT the code does

# 4️⃣ Scalability:
#    This function can now be used in:
#    - Web backends
#    - Mobile apps
#    - ML pipelines
#    - Payment systems

# 👉 This is the shift from "writing scripts" to "building software"
````

fef## Parameters & Arguments

`Parameters` are variables defined inside the function parentheses.
They act as placeholders so the function can work with different values.

`Arguments` are the actual values you pass into the function when calling it.

#### 📌 Why This Matters
Parameters allow your functions to work with **dynamic data**, making them flexible and reusable.

#### Example

```python
def greet_name(name):
    print(f"Hello {name}")
```

Here:
- `name` → Parameter
- It is defined when we create the function.

#### Calling (Running) the Function
```python
greet_name("Sayem")
greet_name("Shovon")
```

Here:
- `Sayem` and `Shovon` → Arguments
- These are the real values passed into the function.


**Default parameter:** A default parameter is a parameter that already has a value assigned in the function definition.

If you don’t provide a value when calling the function, Python uses the default.

```python
# Example
def greet_name(name="Guest"):
    print(f"Hello {name}")

greet_name("Sayem")   # Output: Hello Sayem
greet_name()          # Output: Hello Guest
```

- "Sayem" → overrides the default

- No argument → uses "Guest" as the default

## Return Values

A function can send data back to the main program using return.

📢 Important
- `print()` only displays output
- `return` sends data for further processing

````python
# ============================================================
# EXAMPLE: Understanding `print` vs `return`
# Real Use Case: Student Result Processing System
# ============================================================


# -------------------------------
# ❌ Using print (NOT reusable)
# -------------------------------

def calculate_grade_print(marks):
    grade = "A" if marks >= 80 else "B" # logic
    print("Grade:", grade)


# Output is shown, but cannot be reused
result = calculate_grade_print(85)
print("Stored result:", result)


# ❗ Problem:
# - The grade is displayed
# - But it cannot be stored, compared, or sent elsewhere
````

````python
# ============================================================
# ✅ Using return (Reusable & Powerful)
# ============================================================

def calculate_grade_return(marks):
    grade = "A" if marks >= 80 else "B" # logic
    return grade


# -------------------------------
# Using returned value
# -------------------------------

# student_grade = calculate_grade_return(85)
# print("Grade:", student_grade)

result = calculate_grade_return(85)
print("Stored result:", result)
print("Grade:", result)
````

````python

# ============================================================
# ✅ Why `return` is important
# ============================================================

# 1️⃣ The returned value can be stored in variables
# 2️⃣ It can be used in conditions
# 3️⃣ It can be passed to another function
# 4️⃣ Essential for real-world software logic


# Example: Decision making using returned value
if student_grade == "A":
    print("Eligible for scholarship")
else:
    print("Keep improving!")


# 👉 Rule of Thumb:
# Use `print()` for debugging or displaying info
# Use `return` for building real applications
````

````python
result = calculate_grade_return(85)
print("Stored result:", result)
print("Grade:", result)


def check_scolarship(grade):
  if grade == "A":
      print("Eligible for scholarship")
  else:
      print("Keep improving!")


result = calculate_grade_return(85)
print("Stored result:", result)
print("Grade:", result)
check_scolarship(result)
````

## Keyword Arguments

🏷️ Keyword Arguments

Explicitly specify parameter names — order does not matter.

These features make functions easier to use and harder to misuse.

````python
def calculate_final_price(price, tax_rate):
    total = price + price*tax_rate
    print(f"Total: {total}")
    return total
````

````python
# postion matters -> positional arguments
calculate_final_price(0.15, 3000)   # wrong argument order-> 0.15 (tax_rate) should be the 2nd argument
calculate_final_price(3000, 0.15)   # right argument order
````

````python
# postion doesn't matter  -> keyword arguments
calculate_final_price(price=3000, tax_rate=0.15)
calculate_final_price(tax_rate=0.15, price=3000)
````

## Variable Scope

`Scope` means where a variable can be accessed in your code.


1. Local Scope: A variable created inside a function
```python
def greet():
    message = "Hello"
    print(message)

greet()    ##  works
print(message) ## Error
```
- variable `message` works inside the function
- It does NOT exist outside
- Local variables live only inside the function


2. Global Scope: A variable created outside all functions.

```python
name = "Shovon"

def show():
    print(f"Hello {name}")

show()   ## works
print(name)  ## also works
```

- Variable `name` is accessible both inside and outside of the function
- Global variables can be read inside functions.

#### 🏠 Think of Scope Like Rooms in a House

- **Local Scope**: Variables inside a function (private room)
- **Global Scope**: Variables outside functions (shared space)

Understanding scope prevents **unexpected bugs**.

````python
def greet():
    user = "Sayem"   # local scope
    print(f"Hello, welcome back {user}")
greet()
print(user) # here variable `user` cannot be accessed outside of the function
````

````python
user_name = "Jahid"  # global scope

def greet_v2():
    print(f"Hello, welcome back {user_name}")

greet_v2()
print(user_name) # here variable `user_name` can be accessed outside of the function
````

````python
app_name = "AI-Manager"  # Global variable


def register_user():
    print(f"Thanks for signing up in {app_name}")

register_user()

def login_user():
    username = "Shovon"  # Local variable
    print(f"Logging into {app_name} as {username}")

login_user()

````

## Error Handling with Try & Except
Errors happen. Good software handles them gracefully instead of crashing.

🎯 Why Use Try-Except?
- Protect your program from invalid user input
- Improve user experience
- Essential for production-ready code

````python
def divide(a, b):
    if b == 0:
        return "Division not possible"
    return a/b
````

````python
print(divide(10, 2))
````

````python
print(divide(10, 0))
````

````python
print(divide(10, "a"))
````

````python
# SYNTAX

try:
  pass # implement your logics here
except:
  pass # handle the failure
finally:
  pass # a final block of code that will run
````

````python
def safe_divide(a, b):
    try:
        if b==0 :
            return "Not possible"
        result = a / b
        return result
    except Exception as e:
        return e

print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide(10, "a"))
````

````python
def safe_divide(a, b):
    try:
        result = a / b
        return result
    # custom user friendly messages for each errors (2 except blocks)
    except ZeroDivisionError:
        return "Error: You cannot divide by zero!"
    except TypeError:
        return "Error: Please provide numeric values only."

print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide(10, "a"))
````

````python
def safe_divide(a, b):
    try:
        return a / b
    # a single custom error message for all errors
    except Exception:
        return "Error: Something went wrong. Please check your input."

print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide(10, "a"))
````
